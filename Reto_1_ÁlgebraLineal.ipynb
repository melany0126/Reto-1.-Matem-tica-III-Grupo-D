```python
# librería sympy instalada
from sympy import Matrix, Rational
```

###Problema 1: Ingeniería — Fuerzas en una Armadura

```python
# INGRESAR AMBAS MATRICES
# ingresamos los coeficientes del sistema como una matriz
# cada fila es una ecuación, cada columna es una variable (x1, x2, x3)
A1 = Matrix([
    [2, 1, 0],
    [1, 3, 1],
    [0, 1, 2]
])

# ingresamos el vector con las cargas en cada nodo (kN)
b1 = Matrix([10, 20, 15])

print('Matriz A:')
print(A1)
print()
print('Vector b:')
print(b1)
```

```python
# CALCULAR FACTORIZACIÓN LU
# con LUdecomposition() descompone A en dos matrices triangulares L y U
L1, U1, _ = A1.LUdecomposition()

print('Matriz L:')
print(L1)
print()
print('Matriz U:')
print(U1)
```

```python
# VERIFICO QUE L*U=A
# multiplicamos L por U y comprobamos que da la matriz original A
# si da True, la factorizacion esta bien hecha
print('L multiplicado por U es igual a A:', L1 * U1 == A1)
```

```python
# RESOLVEMOS EL SISTEMA
# sustitucion hacia adelante, resolvemos L*y = b
# sympy despeja y ecuacion por ecuacion de arriba hacia abajo
y1 = L1.solve(b1)
print('Vector y (resultado intermedio):')
print(y1)

# sustitucion hacia atras, resolvemos U*x = y
# sympy despeja x ecuacion por ecuacion de abajo hacia arriba
x1 = U1.solve(y1)
print()
print('Fuerzas internas en cada barra (kN):')
print('  x1 =', x1[0], '=', float(x1[0]), 'kN')
print('  x2 =', x1[1], '=', float(x1[1]), 'kN')
print('  x3 =', x1[2], '=', float(x1[2]), 'kN')
```

```python
# VERIFICO SI ESTA CORRECTO
# multiplicamos A por x y comprobamos que da el vector b original
print('A multiplicado por x es igual a b:', A1 * x1 == b1)
```

### Problema 2: Economía - Producción entre Sectores


```python
# INGRESAR AMBAS MATRICES
# uso rotational para que las fracciones sean exactas y no decimales
A2 = Matrix([
    [ Rational(4,5),  Rational(-1,10), Rational(-1,20)],  # fila Agricultura
    [Rational(-1,4),  Rational( 4, 5), Rational(-1,10)],  # fila Industria
    [Rational(-1,10), Rational(-1,10), Rational( 4, 5)]   # fila Servicios
])

# demanda final de los consumidores para cada sector
b2 = Matrix([100, 80, 60])

print('Matriz A:')
print(A2)
print('Vector b (demanda final en millones USD):')
print(b2)
```

```python
# CALCULAR FACTORIZACIÓN LU
# con LUdecomposition() descompone A en dos matrices triangulares L y U
L2, U2, _ = A2.LUdecomposition()

print('Matriz L:')
print(L2)
print()
print('Matriz U:')
print(U2)
```

```python
# VERIFICO QUE L*U=A
print('L multiplicado por U es igual a A:', L2 * U2 == A2)
```

```python
# RESOLVEMOS EL SISTEMA

# sustitucion hacia adelante
y2 = L2.solve(b2)
print('Vector y:')
print(y2)

# sustitucion hacia atras
x2 = U2.solve(y2)
print()
print('Produccion total que necesita cada sector (millones USD):')
sectores = ['Agricultura', 'Industria', 'Servicios']
for nombre, valor in zip(sectores, x2):
    print(f'  {nombre}: {float(valor):.2f} M USD')
```

```python
# VERIFICO SI ESTA CORRECTO
print('A multiplicado por x es igual a b:', A2 * x2 == b2)
# comparamos produccion total vs demanda final
demanda = [100, 80, 60]
print()
print('Sector        | Demanda final | Produccion total | Va como insumo')
print('-' * 65)
for nombre, val, dem in zip(sectores, x2, demanda):
    total = float(val)
    insumo = total - dem
    print(f'{nombre:<13} | {dem:>13} | {total:>16.2f} | {insumo:>14.2f}')
```

###Problema 3: Medio Ambiente - Contaminante en un Río

```python
# INGRESAR AMBAS MATRICES
# ingresamos los coeficientes del sistema de balance de masa
A3 = Matrix([
    [ 3, -1,  0],
    [-1,  4, -1],
    [ 0, -1,  3]
])

# cargas externas de contaminante en cada zona (g/s)
b3 = Matrix([20, 10, 15])

print('Matriz A:')
print(A3)
print('Vector b (cargas externas en g/s):')
print(b3)
```

```python
# CALCULAR FACTORIZACIÓN LU
L3, U3, _ = A3.LUdecomposition()

print('Matriz L:')
print(L3)
print()
print('Matriz U:')
print(U3)

# algo importante: los multiplicadores de L son negativos porque A tiene -1
# fuera de la diagonal. eso es normal y no es un error.
```

```python
# VERIFICO QUE L*U=A
print('L multiplicado por U es igual a A:', L3 * U3 == A3)
```

```python
# RESOLVEMOS EL SISTEMA
# sustitucion hacia adelante
y3 = L3.solve(b3)
print('Vector y:')
print(y3)

# sustitucion hacia atras
c3 = U3.solve(y3)
print()
print('Concentracion de contaminante en cada zona (g/m3):')
for i, valor in enumerate(c3, 1):
    print(f'  Zona {i}: {valor} = {float(valor):.4f} g/m3')
```

```python
# VERIFICO SI ESTA CORRECTO
# comprobamos que A*c da exactamente b
print('A multiplicado por c es igual a b:', A3 * c3 == b3)

# comparamos con un limite normativo de ejemplo
limite = 8  # g/m3, valor tipico para agua de riego
print()
print(f'Limite normativo: {limite} g/m3')
print()
for i, valor in enumerate(c3, 1):
    estado = 'SUPERA EL LIMITE' if float(valor) > limite else 'dentro del limite'
    print(f'  Zona {i}: {float(valor):.4f} g/m3  ->  {estado}')
```
